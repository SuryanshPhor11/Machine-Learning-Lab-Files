# Ques 11 Create a Pandas DataFrame using a dictionary of lists.

```python
import pandas as pd

# Create a dictionary of lists
data = {
    "Name": ["Alice", "Bob", "Charlie", "David"],
    "Age": [25, 30, 35, 40],
    "City": ["New York", "Los Angeles", "Chicago", "Houston"]
}

# Create DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
print(df)
```

# Ques 12 Filter rows in a DataFrame where a specific column value is greater than a threshold.

```python
import pandas as pd

# Sample DataFrame
data = {
    "Name": ["Alice", "Bob", "Charlie", "David"],
    "Age": [25, 30, 35, 40],
    "City": ["New York", "Los Angeles", "Chicago", "Houston"]
}

df = pd.DataFrame(data)

# Filter rows where Age > 30
filtered_df = df[df["Age"] > 30]

print("Filtered DataFrame:\n", filtered_df)
```

# Ques 13 Add a new column to a DataFrame based on an existing columnâ€™s values. 

```python
import pandas as pd

# Sample DataFrame
data = {
    "Name": ["Alice", "Bob", "Charlie", "David"],
    "Age": [25, 30, 35, 40]
}

df = pd.DataFrame(data)

# Add a new column 'Age Group' based on 'Age'
df["Age Group"] = df["Age"].apply(lambda x: "Young" if x < 30 else "Adult")

print(df)
```

# Ques 14 Group data by a categorical column and calculate the average of numeric columns.

```python
import pandas as pd

# Sample DataFrame
data = {
    "Department": ["HR", "IT", "HR", "IT", "Finance", "Finance"],
    "Employee": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank"],
    "Salary": [50000, 60000, 55000, 65000, 70000, 72000],
    "Age": [25, 30, 28, 35, 40, 38]
}

df = pd.DataFrame(data)

# Group by 'Department' and calculate mean of numeric columns
grouped = df.groupby("Department")[["Salary", "Age"]].mean()
print(grouped)
```

# Ques 15 Plot a line graph using Matplotlib for a dataset column over time.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample dataset
data = {
    "Date": pd.date_range(start="2025-10-01", periods=7, freq="D"),
    "Sales": [150, 200, 180, 220, 210, 230, 250]
}

df = pd.DataFrame(data)

# Plot a line graph
plt.figure(figsize=(8,5))
plt.plot(df["Date"], df["Sales"], marker='o', linestyle='-', color='b')

# Add titles and labels
plt.title("Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.grid(True)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

# Show the plot
plt.tight_layout()
plt.show()
```

# Ques 16 Create a bar chart showing frequency of categories in a dataset column.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample dataset
data = {
    "Department": ["HR", "IT", "HR", "Finance", "IT", "Finance", "HR", "IT"]
}

df = pd.DataFrame(data)

# Count frequency of each category
category_counts = df["Department"].value_counts()

# Plot a bar chart
plt.figure(figsize=(6,4))
plt.bar(category_counts.index, category_counts.values, color='skyblue')

# Add titles and labels
plt.title("Frequency of Departments")
plt.xlabel("Department")
plt.ylabel("Count")
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

# Ques 17 Generate a histogram of a numerical column from a dataset using Matplotlib.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample dataset
data = {
    "Age": [22, 25, 27, 30, 22, 24, 29, 31, 28, 26, 27, 30]
}

df = pd.DataFrame(data)

# Plot histogram
plt.figure(figsize=(7,5))
plt.hist(df["Age"], bins=5, color='skyblue', edgecolor='black')

# Add titles and labels
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

# Ques 18 Plot a scatter plot of two numerical variables using Matplotlib.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample dataset
data = {
    "Hours_Studied": [2, 3, 4, 5, 6, 7, 8],
    "Scores": [50, 55, 60, 65, 70, 75, 80]
}

df = pd.DataFrame(data)

# Plot scatter plot
plt.figure(figsize=(7,5))
plt.scatter(df["Hours_Studied"], df["Scores"], color='green', s=100, alpha=0.7)

# Add titles and labels
plt.title("Scores vs Hours Studied")
plt.xlabel("Hours Studied")
plt.ylabel("Scores")
plt.grid(True)

plt.show()
```

# Ques 19 Customize your plot: add title, axis labels, and legend using Matplotlib.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample dataset
data = {
    "Months": ["Jan", "Feb", "Mar", "Apr", "May"],
    "Sales_2024": [150, 200, 180, 220, 210],
    "Sales_2025": [180, 210, 190, 230, 220]
}

df = pd.DataFrame(data)

# Plot multiple lines
plt.figure(figsize=(8,5))
plt.plot(df["Months"], df["Sales_2024"], marker='o', linestyle='-', color='blue', label='Sales 2024')
plt.plot(df["Months"], df["Sales_2025"], marker='s', linestyle='--', color='red', label='Sales 2025')

# Add title and axis labels
plt.title("Monthly Sales Comparison")
plt.xlabel("Month")
plt.ylabel("Sales (in $1000)")

# Add legend
plt.legend()

# Add grid
plt.grid(True, linestyle='--', alpha=0.7)

plt.show()
```

# Ques 20 Save a Matplotlib plot to a file (e.g., PNG or JPG) and display it inline in Jupyter Notebook.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample dataset
data = {
    "Months": ["Jan", "Feb", "Mar", "Apr", "May"],
    "Sales": [150, 200, 180, 220, 210]
}

df = pd.DataFrame(data)

# Plot the data
plt.figure(figsize=(7,5))
plt.plot(df["Months"], df["Sales"], marker='o', linestyle='-', color='blue', label='Sales')

# Add title and labels
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

# Save the plot to a file (PNG or JPG)
plt.savefig("monthly_sales.png")   # you can also use "monthly_sales.jpg"

# Display the plot inline
plt.show()
```
